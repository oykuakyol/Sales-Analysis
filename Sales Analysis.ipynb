```python
import pandas as pd

data = {
    "order_id": [1,2,3,4,5,6],
    "product": ["Laptop", "Mouse", "Keyboard", "Laptop", "Mouse", "Monitor"],
    "price": [20000, 500, 700, 20000, 500, 8000],
    "quantity": [1, 2, 1, 1, 3, 1],
    "date": ["2024-01-01", "2024-01-02", "2024-01-03",
             "2024-01-05", "2024-01-06", "2024-01-07"]
}

df = pd.DataFrame(data)
df
```

```python
df["total_sales"] = df["price"] * df["quantity"]
df
```

```python
df.groupby("product")["total_sales"].sum()
```

```python
df.groupby("date")["total_sales"].sum().sort_values(ascending=False)
```

```python
df["total_sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("product")["total_sales"].sum().plot(kind="bar")

plt.title("Ürünlere Göre Satış")
plt.xlabel("Ürün")
plt.ylabel("Toplam Satış")
plt.show()
```
